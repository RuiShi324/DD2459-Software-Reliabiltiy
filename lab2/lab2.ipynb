```python
def binary_sole(arr,key):
    low=0
    high=len(arr)-1
    mid=0
    
    while low<=high:
        
        mid=(low+high)//2
        
        if arr[mid]<key:
            low=mid+1
            
        elif arr[mid]>key:
            high=mid-1
            
        else:
            return mid
        
    return -1
```

### This function sorts the list with bubble sort, and it generates a index for the list at the same time.


```python
def bubblesort(arr):
    n = len(arr)
    arr_sort = []
    for k in range(0,n):
        arr_sort.append(arr[k])
    arr_index = list(range(0, len(arr_sort)))
    for i in range(n-1):

        for j in range(0, n-i-1):
 
            if arr_sort[j] > arr_sort[j + 1]:
                temp=arr_sort[j]
                arr_sort[j]=arr_sort[j+1]
                arr_sort[j+1]=temp

                index_temp=arr_index[j]
                arr_index[j]=arr_index[j+1]
                arr_index[j+1]=index_temp
    
    return arr_sort, arr_index
    
```

### The following two functions can find the start and the end index of the sorted list.

```python
def binary_left(arr , key):
    low = 0
    high = len(arr) - 1
    while low < high:
        mid = (low + high) // 2
        if arr[mid] >= key:
            high = mid
        else:
            low = mid + 1
    if arr[high] == key:
        return high
    return -1

```

```python
def binary_right(arr , key):
    low = 0
    high = len(arr) - 1
    while low < high:
        mid = (low + high) // 2 + 1
        if arr[mid] <= key:
            low = mid
        else:
            high = mid - 1
    if arr[high] == key:
        return high
    return -1
```

### Then combine them, this function could find all the index of the number.

```python
def binary_search(arr, key):
    arr_sort, arr_index = bubblesort(arr)
    left = binary_left(arr_sort, key)
    right = binary_right(arr_sort, key)
    if left == -1:
        print("\nelement ", key, "is not found")
        return []
    else:
        print("\nElemnet",key,"is founded, the index is",arr_index[left:right+1:1])
        return arr_index[left:right+1:1]
```
